```python
import pandas as pd
df1 = pd.read_csv(r'C:\Users\User\Desktop\pythonfoundation\data\addresses.csv')
df1['Add']=df1.Address1 + ', ' + df1.Address3 + ', ' + df1.Address4 + ', ' + df1.Address5
df1.append(df1.Add)
df1.to_csv('FullAdd.csv')
df1.head()
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
 
geolocator = Nominatim(user_agent="my_request")
 
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
 
df1['location'] = df1['Add'].apply(geocode)
df1['Lat'] = df1['location'].apply(lambda x: x.latitude if x else None)
df1['Lon'] = df1['location'].apply(lambda x: x.longitude if x else None)

df1.dropna(inplace=True)
df1.to_csv('FullAdd.csv')
df1.head()
```
